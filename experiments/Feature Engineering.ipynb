```python
import pandas as pd
import numpy as np
```

```python
lyrics_df = pd.read_csv("Song Lyrics After Cleaning.csv", usecols=['artist', 'song name', 'lyrics', 'length', 'Mood'], index_col='song name')
lyrics_df
```

```python
lyrics_df['length_log'] = np.log(lyrics_df['length'])
lyrics_df = lyrics_df.drop('length', 1)
```

```python
lyrics_df['artist'] = lyrics_df['artist'].astype('category').cat.codes
lyrics_df
```

```python
lyrics_df['Mood'].value_counts()
```

```python
requires_n = lyrics_df['Mood'].value_counts().min()

negative_mood = lyrics_df[lyrics_df['Mood'] == 0].sample(n=requires_n)
positive_mood = lyrics_df[lyrics_df['Mood'] == 1]

ML_data = pd.concat([negative_mood, positive_mood])

# The frac keyword argument specifies the fraction of rows to return in the random sample, 
# so frac=1 means to return all rows (in random order).
ML_data = ML_data.sample(frac=1)
```

```python
ML_data
```

```python
ML_data.to_csv("ML data.csv")
```
